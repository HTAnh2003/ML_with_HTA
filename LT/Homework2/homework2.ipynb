### Bài 1 :
Bạn viết một chương trình máy tính    
Dữ liệu đầu vào của người dùng được giả định là một công thức bao gồm một số , một toán tử ( ít nhât là toán tử + và - ) và một số khác , được phân tách bằng khoảng trắng (ví dụ : 1 + 1 ) . Phân tách đầu vào của người dùng bằng `str.split()` và kiểm tra xem danh sách kết quả có hợp lệ hay không :
- Nếu đầu vào không bao gồm 3 phần tử , hãy đưa ra `FormulaError`.
- Hãy chuyển đổi đầu vào thứ nhất và thứ ba thành float ( như vậy : float_value = float(str_value)). Bắt bất kỳ ValueError nào xảy ra và thay vào đó đưa ra `FormulaError`.
- Nếu đầu vào thứ hai không phải là `+` hoặc `-` , hãy tạo lại `FormulaError`.
Nếu dữ liệu nhập vào hợp lệ , thực hiện phép tính và in ra kết quả. Sau đó , người dùng sẽ được nhắc cung cấp thông tin đầu vào mới,... cho đến khi người dùng thoát

```python
class FormulaError(Exception):
    pass

def compute(str):
    print(str)
    try :
        if len(str.split()) == 3 :
            number1, operator, number2 = str.split()
        else :
            raise FormulaError("FormulaError")
        
        number1 = float(number1)
        number2 = float(number2)
        
        if operator == "+" or operator == "-" :
            if operator == "+" :
                return number1 + number2
            else :
                return number1 - number2
        else :
            raise FormulaError("FormulaError")
    except ValueError :
        raise FormulaError("FormulaError")


# Chạy đến khi nguời dùng thoát
while True:
    input_str = input("Nhập công thức (số + hoặc - số) hoặc gõ 'exit' để thoát: ")
    
    if input_str.lower() == 'exit':
        break
    
    try:
        result = compute(input_str)
        print("Kết quả:", result)
    except FormulaError as e:
        print(e)
        continue
```

```python
class Person:
    def __init__(self, name, address) :
        self.name = name
        self.address = address

    def getName(self):
        return self.name

    def getAddress(self):
        return self.address

    def setAddress(self, address):
        self.address = address

    def __str__(self):
        # Format: name : name | address : address
        return "name : " + self.name + " | address : " + self.address
```

```python
class Student(Person):
    def __init__(self, name, address, program, year, fee):
        super().__init__(name, address)
        self.program = program
        self.year = year
        self.fee = fee

    def getProgram(self):
        return self.program

    def setProgram(self, program):
        self.program = program

    def getYear(self):
        return self.year

    def setYear(self, year):
        self.year = year

    def getFee(self):
        return self.fee

    def setFee(self, fee):
        self.fee = fee

    def __str__(self):
        return "name : " + self.name + " | address : " + self.address + " | program : " + self.program + " | year : " + str(self.year) + " | fee : " + str(self.fee)
```

```python
class Staff(Person):
    def __init__(self, name, address, school, pay):
        super().__init__(name, address)
        self.school = school
        self.pay = pay

    def getSchool(self):
        return self.school

    def setSchool(self, school):
        self.school = school

    def getPay(self):
        return self.pay

    def setPay(self, pay):
        self.pay = pay

    def __str__(self):
        return "name : " + self.name + " | address : " + self.address + " | school : " + self.school + " | pay : " + str(self.pay)
```

```python
# Viết chương trình thực hiện menu theo 3 yêu cầu sau :
# 1.Thêm vào dữ liệu Student và Staff
# 2.Viết funtion tìm kiếm theo tên Student và in ra kết quả 
# 3.Thoát khỏi chương trình 

def add_student(students):
    name = input("Nhập tên sinh viên: ")
    address = input("Nhập địa chỉ: ")
    program = input("Nhập chương trình: ")
    year = input("Nhập năm học: ")
    fee = input("Nhập học phí: ")
    student = Student(name, address, program, year, fee)
    students.append(student)

def add_staff(staffs):
    name = input("Nhập tên nhân viên: ")
    address = input("Nhập địa chỉ: ")
    school = input("Nhập trường: ")
    pay = input("Nhập lương: ")
    staff = Staff(name, address, school, pay)
    staffs.append(staff)

def search_student(students):
    name = input("Nhập tên sinh viên cần tìm: ")
    found_students = [student for student in students if student.getName().lower() == name.lower()]
    if found_students:
        for student in found_students:
            print(student)
    else:
        print("Không tìm thấy sinh viên với tên đó.")

def main():
    students = []
    staffs = []

    while True:
        print("\n1. Thêm Student và Staff ")
        print("2. Tìm kiếm Sinh viên theo tên")
        print("3. Thoát")

        choice = input("Nhập lựa chọn của bạn (1-3): ")

        if choice == '1':
            print("\n1. Thêm Student")
            print("2. Thêm Staff")
            choice = input("Nhập lựa chọn của bạn (1-2): ")
            if choice == '1':
                add_student(students)
            elif choice == '2':
                add_staff(staffs)
            else:
                print("Lựa chọn không hợp lệ. Vui lòng nhập lại!")
        elif choice == '2':
            search_student(students)
        elif choice == '3':
            print("Thoát chương trình.")
            break
        else:
            print("Lựa chọn không hợp lệ. Vui lòng nhập lại!")

if __name__ == "__main__":
    main()
```

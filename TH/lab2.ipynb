## **Bài 1** : 
Một giám đốc nhân sự mong muốn so sánh tính hiệu quả của hai phương pháp huấn luyện các nhân viên công nghiệp nhằm thực hiện một hoạt động lắp ráp nào đó. Một số lượng nhân viên được chia thành hai nhóm bằng nhau, n hóm thứ nhất nhận được phương pháp huấn luyện 1 và nhóm thứ hai được huấn luyện bằng phương pháp 2. Mỗi nhóm sẽ thực hiện hoạt động lắp ráp này, và độ dài của thời gian lắp ráp sẽ được ghi nhận. Người ta kỳ vọng rằng các đại lượng cho cả hai nhóm sẽ có một khoảng xấp xỉ 8 phút. Để cho ước lượng về sự khác biệt về thời gian trung bình để lắp ráp chính xác trong giới hạn 1 phút với xác suất bằng với 0.95, thì cần phải đưa bao nhiêu công nhân vào mỗi nhóm huấn luyện

### Bài Toán
So sánh hiệu quả của hai phương pháp huấn luyện nhân viên công nghiệp trong một hoạt động lắp ráp, và muốn ước lượng sự khác biệt về thời gian trung bình để lắp ráp chính xác trong giới hạn 1 phút với xác suất 0.95.

### Giả Định
1. Độ lệch chuẩn ($ \sigma $) cho mỗi phương pháp huấn luyện là tương đương.
2. Độ biến thiên (range = max - min) của từng phương pháp lắp ráp là khoảng 8 phút, và giả định rằng này xấp xỉ bằng với 4$ \sigma $, từ đó $ \sigma \approx 2 $ phút.

### Phương Pháp Tính Toán
Sử dụng công thức sau để xác định kích thước mẫu cho mỗi nhóm:

$$ n = \frac{(Z_{\alpha/2}^2 \times 2\sigma^2)}{d^2} $$

trong đó:
- $ Z_{\alpha/2} $ là giá trị Z cho độ tin cậy 0.95, tương ứng là 1.96 (vì 1 - (1 - 0.95)/2 = 0.975 và Z cho 0.975 là 1.96).
- $ \sigma $ là độ lệch chuẩn, ở đây giả định là 2 phút.
- $ d $ là độ chính xác mong muốn, ở đây là 1 phút.

### Tính Toán
Thay thế các giá trị vào công thức:

$$ n = \frac{1.96^2 \times (2 \times 2^2)}{1^2} $$
$$ n = \frac{1.96^2 \times 8}{1} $$
$$ n = \frac{3.8416 \times 8}{1} $$
$$ n = 30.7328 $$

Vì kích thước mẫu không thể là số thập phân, làm tròn lên và nhận được $ n = 31 $.

### Kết Luận
Mỗi nhóm trong nghiên cứu cần có khoảng 31 công nhân để đạt được độ chính xác 1 phút với xác suất 0.95 trong việc ước lượng sự khác biệt về thời gian lắp ráp giữa hai phương pháp huấn luyện.

## **Bài 2**: 
Một công ty kiểm toán mong muốn ước lượng sai số trung bình mỗi tài khoản trong các khoản phải thu cho một công ty cung cấp hệ thống ống nước chính xác trong giới hạn $20 với xác suất bằng 0.99. Một mẫu nhỏ trước đó gợi ý rằng sai số mỗi tài khoản sở hữu một độ lệch chuẩn xấp xỉ bằng với $58. Nếu công ty kiểm toán đó mong muốn ước lượng sai số trung bình mỗi tài khoản chính xác trong giới hạn $20, thì có bao nhiêu tài khoản ắt sẽ phải được chọn mẫu? Mẫu này phải sở hữu (các) thuộc tính nào

Để ước lượng sai số trung bình mỗi tài khoản trong các khoản phải thu với độ chính xác nhất định, ta cần sử dụng công thức của kích thước mẫu cho ước lượng trung bình với một độ tin cậy cụ thể. Trong trường hợp này, công thức đó sẽ dựa vào giá trị Z của phân phối chuẩn tương ứng với độ tin cậy 0.99 và độ lệch chuẩn của quần thể.

Công thức để xác định kích thước mẫu $ n $ là:

$$ n = \left( \frac{Z \times \sigma}{E} \right)^2 $$

Trong đó:
- $ Z $ là giá trị Z tương ứng với độ tin cậy mong muốn (đối với độ tin cậy 0.99, $ Z $ khoảng 2.576).
- $ \sigma $ là độ lệch chuẩn của quần thể (trong trường hợp này là $58).
- $ E $ là sai số tối đa cho phép (trong trường hợp này là $20).

Thay thế các giá trị vào công thức:

$$ n = \left( \frac{2.576 \times 58}{20} \right)^2 $$

Tính toán kích thước mẫu $ n $:

$$ n = \left( \frac{149.408}{20} \right)^2 $$
$$ n = 7.4704^2 $$
$$ n = 55.808 $$

Vì kích thước mẫu không thể là một số thập phân, chúng ta cần làm tròn lên đến số nguyên gần nhất, nên $ n = 56 $.

Vậy, công ty kiểm toán cần chọn mẫu ít nhất 56 tài khoản để ước lượng sai số trung bình mỗi tài khoản chính xác trong giới hạn $20 với xác suất 0.99.

Từ đồ thị thì ta có thể thấy các bạn nữ thường phần đọc hiểu sẽ cao hơn các bạn nam, có thể do các bạn nữ sẽ kiên nhẫn, đọc kĩ hơn các bạn nam

```python
#b. Tự viết các hàm tính Trung binh (mean), Độ lệch chuẩn (standard deviation) và khoảng dữ liệu (range) và áp dụng các hàm này với cột dữ liệu pv1MATH.

# Hàm tính trung bình
def calculate_mean(data):
    return sum(data) / len(data)

def calculate_standard_deviation(data):
    mean = calculate_mean(data)
    variance = sum([(x - mean) ** 2 for x in data]) / len(data)
    return variance ** 0.5

def calculate_range(data):
    return max(data) - min(data)

# Giả sử df là DataFrame của bạn và 'pv1MATH' là cột dữ liệu
pv1MATH_data = df['PV1MATH']

# Tính toán
mean_pv1MATH = calculate_mean(pv1MATH_data)
std_dev_pv1MATH = calculate_standard_deviation(pv1MATH_data)
range_pv1MATH = calculate_range(pv1MATH_data)

print("Trung bình pv1MATH:", mean_pv1MATH)
print("Độ lệch chuẩn pv1MATH:", std_dev_pv1MATH)
print("Khoảng dữ liệu pv1MATH:", range_pv1MATH)
```

## Bài 4:
Một công ty muốn dự đoán số phút gọi vào tổng đài dịch vụ của khách dựa vào số linh kiện cần sửa chữa (Units), hãy tính βo và β1 và xây dựng mô hình hồi quy theo mô hình
sau:

    Số phút = βo + β1Units
Dữ liệu được cho như bảng dưới đây :



Viết chương trình xây dựng một ứng dụng với chức năng nhập vào một dữ liệu và đưa ra dự đoán số phút gọi vào dịch vụ tổng đài.


```python
import numpy as np
import pandas as pd
data = np.array([[49,3],[64,4],[23,1],[29,2],[87,5],[96,6],[74,4],[145,9],[154,10],[166,10],[97,6],[109,7],[111,8],[149,9]])
df = pd.DataFrame(data, columns=['Số phút', 'Units'])
df
```

```python
# Tính βo và β1
x = df['Units']
y = df['Số phút']

# Sử dụng thư viện sklearn
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(df[['Units']], df[['Số phút']])
print("Hàm hồi quy theo sklearn: y = {} + {}x ".format(model.intercept_[0], model.coef_[0][0]))

# Sử dụng thống kê để tính βo và β1 thủ công
n = len(x)
x_mean = np.mean(x)
y_mean = np.mean(y)
xy_mean = np.mean(x*y)
x2_mean = np.mean(x**2)
beta1 = (xy_mean - x_mean*y_mean)/(x2_mean - x_mean**2)
beta0 = y_mean - beta1*x_mean
print("Hàm hồi quy theo thống kê: y = {} + {}x ".format(beta0, beta1))

# Vẽ đồ thị
import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.plot(x, model.predict(df[['Units']]), color='red')
plt.show()




```

```python
# Dư đoán số phút gọi vào tổng đài dịch vụ
units = int(input("Nhập số linh kiện cần sửa chữa: "))
print("Dự đoán số phút gọi vào tổng đài dịch vụ: ", model.predict([[units]])[0][0])# Dư đoán số phút gọi vào tổng đài dịch vụ
print("Dự đoán số phút gọi vào tổng đài dịch vụ: ", beta0 + beta1*units)
```

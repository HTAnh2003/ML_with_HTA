Bảng phân phối xác suất của X :

```python
def expected_value_discrete(p, x):
    expected_value = 0
    for i in range(len(p)):
        expected_value += p[i] * x[i]
    return expected_value

from scipy.integrate import quad # tính tích phân

def expected_value_continuous(fx, lower_bound, upper_bound):
    """
    Tính kỳ vọng cho biến ngẫu nhiên liên tục.
    :param fx: Hàm mật độ xác suất của biến ngẫu nhiên liên tục.
    :param lower_bound: Giới hạn dưới của tích phân.
    :param upper_bound: Giới hạn trên của tích phân.
    :return: Giá trị kỳ vọng.
    :
    """
    expected_value = quad(lambda x: x * fx(x), lower_bound, upper_bound)[0]
    return expected_value
```

```python
def variance_discrete (p, x):
    '''
    Tính phương sai cho biến ngẫu nhiên rời rạc
    :param p: List xác suất
    :param x: List các giá trị
    :return: Giá trị phương sai
    '''
    expected_value = expected_value_discrete(p, x)
    variance = 0
    for i in range(len(p)):
        variance += p[i] * (x[i] - expected_value) ** 2
    return variance

def variance_continuous(fx, lower_bound, upper_bound):
    """
    Tính phương sai cho biến ngẫu nhiên liên tục.
    :param fx: Hàm xác suất
    :param lower_bound: Giới hạn dưới của tích phân.
    :param upper_bound: Giới hạn trên của tích phân.
    :return: Giá trị phương sai.
    """
    expected_value = expected_value_continuous(fx, lower_bound, upper_bound)
    return (quad(lambda x: x ** 2 * fx(x), lower_bound, upper_bound)[0])  -  (expected_value ** 2)
```

```python
# Tinh do lech chuan
def standard_variance_discrete(p, x):
    return variance_discrete(p, x) ** .5

def standard_variance_continuous(pdf, lower_bound, upper_bound):
    return variance_continuous(pdf, lower_bound, upper_bound) ** .5
```

**Bài 6** : Một cuộc điều tra 100 đại lý mua hàng tạo ra một ước lượng về tỷ lệ những người
bán buôn ống nước polyvinyl mà có kế hoạch gia tăng sự mua hàng của mình trong
năm tới. Biên sai số, 0.096, là tương đối lớn. Giả định rằng tổ chức tiếp thị tiến hành
cuộc điều tra này được yêu cầu phải thực hiện một cuộc điều tra mới và đạt được
một ước lượng chính xác trong giới hạn 0.04 với xác suất bằng với 0.90. Xấp xỉ có
bao nhiêu nhà bán buôn ắt đã phải tính đến trong cuộc điều tra này?



Để xác định kích thước mẫu cần thiết cho cuộc điều tra mới sao cho độ chính xác của ước lượng nằm trong giới hạn cho trước với xác suất nhất định, chúng ta có thể sử dụng công thức cho kích thước mẫu trong ước lượng tỷ lệ dân số:

$$ n = \left( \frac{Z^2 \times p \times (1 - p)}{E^2} \right) $$

Trong đó:
- $ n $ là kích thước mẫu cần thiết.
- $ Z $ là giá trị Z tương ứng với xác suất mong muốn (trong trường hợp này là 0.90).
- $ p $ là ước lượng tỷ lệ trong dân số (nếu không biết, chúng ta thường sử dụng $ p = 0.5 $ vì nó cung cấp kích thước mẫu lớn nhất).
- $ E $ là biên sai số mong muốn (trong trường hợp này là 0.04).

Đầu tiên, chúng ta cần xác định giá trị của $ Z $. Đối với xác suất 0.90, giá trị Z tương ứng (được tìm trong bảng phân phối chuẩn hoặc qua một công cụ tính toán) là khoảng 1.645.

Giả sử $ p = 0.5 $ để tối đa hóa kích thước mẫu, và $ E = 0.04 $, chúng ta có:

$$ n = \left( \frac{1.645^2 \times 0.5 \times 0.5}{0.04^2} \right) $$
$$ n = 422.81640624999994 $$

Do đó, kích thước mẫu cần thiết cho cuộc điều tra mới là khoảng 422 đại lý mua hàng để đạt được độ chính xác 0.04 với xác suất 0.90. Trong thực tế, chúng ta thường làm tròn lên, nên có thể cần khoảng 423 đại lý.

### Bài Toán
So sánh hiệu quả của hai phương pháp huấn luyện nhân viên công nghiệp trong một hoạt động lắp ráp, và muốn ước lượng sự khác biệt về thời gian trung bình để lắp ráp chính xác trong giới hạn 1 phút với xác suất 0.95.

### Giả Định
1. Độ lệch chuẩn ($ \sigma $) cho mỗi phương pháp huấn luyện là tương đương.
2. Độ biến thiên (range = max - min) của từng phương pháp lắp ráp là khoảng 8 phút, và giả định rằng này xấp xỉ bằng với 4$ \sigma $, từ đó $ \sigma \approx 2 $ phút.

### Phương Pháp Tính Toán
Sử dụng công thức sau để xác định kích thước mẫu cho mỗi nhóm:

$$ n = \frac{(Z_{\alpha/2}^2 \times 2\sigma^2)}{d^2} $$

trong đó:
- $ Z_{\alpha/2} $ là giá trị Z cho độ tin cậy 0.95, tương ứng là 1.96 (vì 1 - (1 - 0.95)/2 = 0.975 và Z cho 0.975 là 1.96).
- $ \sigma $ là độ lệch chuẩn, ở đây giả định là 2 phút.
- $ d $ là độ chính xác mong muốn, ở đây là 1 phút.

### Tính Toán
Thay thế các giá trị vào công thức:

$$ n = \frac{1.96^2 \times (2 \times 2^2)}{1^2} $$
$$ n = \frac{1.96^2 \times 8}{1} $$
$$ n = \frac{3.8416 \times 8}{1} $$
$$ n = 30.7328 $$

Vì kích thước mẫu không thể là số thập phân, làm tròn lên và nhận được $ n = 31 $.

### Kết Luận
Mỗi nhóm trong nghiên cứu cần có khoảng 31 công nhân để đạt được độ chính xác 1 phút với xác suất 0.95 trong việc ước lượng sự khác biệt về thời gian lắp ráp giữa hai phương pháp huấn luyện.

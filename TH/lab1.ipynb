```python
def expected_value_discrete(p, x):
    expected_value = 0
    for i in range(len(p)):
        expected_value += p[i] * x[i]
    return expected_value

from scipy.integrate import quad # tnh tch phn

def expected_value_continuous(fx, lower_bound, upper_bound):
    """
    Tnh k vng cho bin ngu nhin lin tc.
    :param fx: Hm mt  xc sut ca bin ngu nhin lin tc.
    :param lower_bound: Gii hn di ca tch phn.
    :param upper_bound: Gii hn trn ca tch phn.
    :return: Gi tr k vng.
    :
    """
    expected_value = quad(lambda x: x * fx(x), lower_bound, upper_bound)[0]
    return expected_value
```

```python
def variance_discrete (p, x):
    '''
    Tnh phng sai cho bin ngu nhin ri rc
    :param p: List xc sut
    :param x: List cc gi tr
    :return: Gi tr phng sai
    '''
    expected_value = expected_value_discrete(p, x)
    variance = 0
    for i in range(len(p)):
        variance += p[i] * (x[i] - expected_value) ** 2
    return variance

def variance_continuous(fx, lower_bound, upper_bound):
    """
    Tnh phng sai cho bin ngu nhin lin tc.
    :param fx: Hm xc sut
    :param lower_bound: Gii hn di ca tch phn.
    :param upper_bound: Gii hn trn ca tch phn.
    :return: Gi tr phng sai.
    """
    expected_value = expected_value_continuous(fx, lower_bound, upper_bound)
    return (quad(lambda x: x ** 2 * fx(x), lower_bound, upper_bound)[0])  -  (expected_value ** 2)
```

```python
# Tinh do lech chuan
def standard_variance_discrete(p, x):
    return variance_discrete(p, x) ** .5

def standard_variance_continuous(pdf, lower_bound, upper_bound):
    return variance_continuous(pdf, lower_bound, upper_bound) ** .5
```

### Bi Ton
So snh hiu qu ca hai phng php hun luyn nhn vin cng nghip trong mt hot ng lp rp, v mun c lng s khc bit v thi gian trung bnh  lp rp chnh xc trong gii hn 1 pht vi xc sut 0.95.

### Gi nh
1.  lch chun ($ \sigma $) cho mi phng php hun luyn l tng ng.
2.  bin thin (range = max - min) ca tng phng php lp rp l khong 8 pht, v gi nh rng ny xp x bng vi 4$ \sigma $, t  $ \sigma \approx 2 $ pht.

### Phng Php Tnh Ton
S dng cng thc sau  xc nh kch thc mu cho mi nhm:

$$ n = \frac{(Z_{\alpha/2}^2 \times 2\sigma^2)}{d^2} $$

trong :
- $ Z_{\alpha/2} $ l gi tr Z cho  tin cy 0.95, tng ng l 1.96 (v 1 - (1 - 0.95)/2 = 0.975 v Z cho 0.975 l 1.96).
- $ \sigma $ l  lch chun,  y gi nh l 2 pht.
- $ d $ l  chnh xc mong mun,  y l 1 pht.

### Tnh Ton
Thay th cc gi tr vo cng thc:

$$ n = \frac{1.96^2 \times (2 \times 2^2)}{1^2} $$
$$ n = \frac{1.96^2 \times 8}{1} $$
$$ n = \frac{3.8416 \times 8}{1} $$
$$ n = 30.7328 $$

V kch thc mu khng th l s thp phn, lm trn ln v nhn c $ n = 31 $.

### Kt Lun
Mi nhm trong nghin cu cn c khong 31 cng nhn  t c  chnh xc 1 pht vi xc sut 0.95 trong vic c lng s khc bit v thi gian lp rp gia hai phng php hun luyn.
